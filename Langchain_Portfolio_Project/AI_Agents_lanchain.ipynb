```python

from langchain_classic.agents import AgentExecutor, create_react_agent
from langchain_classic import hub
from langchain_google_genai import ChatGoogleGenerativeAI
import requests
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_core.tools import tool
from langsmith import Client
```

```python
search_tool=DuckDuckGoSearchRun()
search_tool.invoke("Who is known as father of nation")
```

```python
@tool
def fetch_weather_data(city:str)->int:
    """This function is mainly used to fetch the weather of the city or a country"""

    url=f"http://api.weatherapi.com/v1/current.json?key=Your_api_key&q={city}"
    response=requests.get(url)

    return response.json()
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite")
```

```python
result=llm.invoke("hii")
result.text
```

```python
client=Client()
```

```python
prompt = client.pull_prompt('hwchase17/react', dangerously_pull_public_prompt=True)
# React stands for React and Action based agent
```

```python
agent=create_react_agent(
    llm=llm,
    tools=[search_tool,],
    prompt=prompt

)
```

```python
agent_exe=AgentExecutor(
    agent=agent,
    tools=[search_tool,fetch_weather_data],
    verbose=True
)
```

```python
response=agent_exe.invoke({"input":"What is the capital of Madhya Pradesh and give me it's current weather and temperature"})
```
